```python
import pandas as pd                                                          #import the necessary libraries
```

```python
# import the dataset
data_url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
```

```python
users = pd.read_csv(data_url, sep='|', index_col='user_id')                 # 3 assign dataset to users

mean_age_per_occupation = users.groupby('occupation')['age'].mean()            # 4 calculation of mean age per occupation
print("Mean age per occupation:")
print(mean_age_per_occupation)
```

```python
 # 5 Calculation of male ratio per occupation and sorting

print("The Male ratio per occupation and sort it from the most to the least is : ")
print()  # For leaving next line
genderCountPerOccupation = users.groupby(['occupation', 'gender']).gender.count()
totalGenderCount = users.groupby('gender').gender.count()
print((genderCountPerOccupation /
       totalGenderCount).sort_values(ascending=False).xs('M', level=1))
```

```python
# 6 calculation of the minimum and maximum ages for each occupation

min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
print("\nMinimum and maximum ages per occupation:")
print(min_max_age_per_occupation)
```

```python
# 7 calculation of the mean age for each combination of occupation and sex

mean_age_per_occupation_sex = users.groupby(['occupation', 'gender'])['age'].mean()
print("\nMean age for each occupation and sex:")
print(mean_age_per_occupation_sex)
```

```python
# 8 calculation of the percentage of women and men for each occupation

total_count_per_occupation_gender = users.groupby(['occupation', 'gender']).size()
total_count_per_occupation = users.groupby('occupation').size()
gender_percentage = (total_count_per_occupation_gender / total_count_per_occupation) * 100
print("\npercentage of women and men for each occupation:")
print(gender_percentage)
```

#Create a pie chart showing the fraction of all Roman Emperors that were assassinated. Make sure that the pie chart is an even
#circle, labels the categories, and shows the percentage breakdown of the categories.

```python
file_path = '/Users/hp/Desktop/roman-emperor-reigns.csv'
pie_chart = pd.read_csv(file_path)

# calculating the number of assassinated and non-assassinated emperors 
assassinated_emperors = (pie_chart['Cause_of_Death'].str.contains('assassinated', case=False)).sum()
print("\nassassinated_emperors:", assassinated_emperors)
Possi_assassinated_emperors = (pie_chart['Cause_of_Death'].str.contains('possibly assassinated', case=False)).sum()
print("\nPossi_assassinated_emperors:", Possi_assassinated_emperors)

total_assassinated_emperors = assassinated_emperors - Possi_assassinated_emperors
non_assassinated_emperors = len(pie_chart) - total_assassinated_emperors
print("\ntotal_assassinated_emperors:", total_assassinated_emperors,"\nnon_assassinated_emperors:", non_assassinated_emperors)

assassinated_percentage = (total_assassinated_emperors / len(pie_chart)) * 100
non_assassinated_percentage = (non_assassinated_emperors / len(pie_chart)) * 100
labels = ['Assassinated Emperors', 'Non-Assassinated Emperors']
sizes = [assassinated_percentage, non_assassinated_percentage]
colors = ['lightgreen', 'lightblue']
explode = (0.1, 0)  
plt.figure(figsize=(6, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

plt.axis('equal')
plt.title('Assassinated Roman Emperors')
plt.tight_layout()
plt.show()
```
